# Question 6 
Get the binomial distribution with n = 10, p = .7 and k = [0, 1, 2, 3, 4, 5, 6, 7]

Hint - Use stats.binom.pmf() function for this.


```python
n=10
p=.7
k= [0,1,2,3,4,5,6,7]
binomial = stats.binom.pmf(k,n,p) #probabilty mass function (pmf)
print(binomial)
```
