```python
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.
#read the data file
import pandas as pd 
import numpy as np
df = pd.read_csv("Laliga.csv",skiprows=[0]) #including skiprows so that the first row is not considered

for col in df:
  df[col] = df[col].replace('-', 0)  #using replace function to replace hypen with zeros
print(df)
```

```python
#Print all the teams which have started playing between 1930-1980.
 #spliting the string on "-" and displays the starting string till hypen is encountered and comparing between 1930 and 1980
df_deb = df[(pd.to_numeric(df["Debut"].str.split("-").str[0]) >= 1930) & (pd.to_numeric(df["Debut"].str.split("-").str[0]) <= 1980)]  
print(df_deb.loc[:,["Team","Debut"]])
```

```python
#Print the list of teams which came Top 5 in terms of points
df["Points"] = pd.to_numeric(df["Points"]) #converting to numeric as comparsion operation cannot be used on strings
df_sort=df.sort_values(by=["Points"],ascending=False).head(5)  # using Sort_values
print(df_sort.loc[:,["Team","Points"]])
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
#Using the same function, find the team which has maximum and minimum goal difference

def Goal_diff_count(cols) :
    df_new = cols['Team'] ,  (pd.to_numeric(cols["GoalsFor"]) -  pd.to_numeric(cols['GoalsAgainst']))
    return pd.Series(df_new)

#finding the max difference by using apply to call the function Goal_diff_count and sort values 
print(df.apply(Goal_diff_count, axis=1).sort_values(by=[1],ascending=False).head(1))  # finding the max difference

#finding the min difference by using apply to call the function Goal_diff_count and sort values 
print(df.apply(Goal_diff_count, axis=1).sort_values(by=[1],ascending=True).head(1))  # finding the max difference
```

```python
#Create a new column with name “Winning Percent” and append it to the data set
#Percentage of Winning = (GamesWon / GamesPlayed)*100
#If there are any numerical error, replace it with 0%
#Print the top 5 teams which has the highest Winning percentage

#Creating a new Column and replacing numerical error with 0 using fill na
df["WinningPercent"] = ((pd.to_numeric(df["GamesWon"]) / pd.to_numeric(df["GamesPlayed"]))*100).fillna(0).astype(int)
print(df)  # shows the new column WinningPercent appended to the last of the series and fillna is used to replace NaN with 0
```

```python
#printing the top 5 teams
df.sort_values(['WinningPercent'],ascending=False).head(5).groupby('WinningPercent')["WinningPercent","Team"].head(5)
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions (10 points)
df["Points"] = pd.to_numeric(df["Points"])
df.groupby(["BestPosition"]).agg({"Points":["sum"]})
```
