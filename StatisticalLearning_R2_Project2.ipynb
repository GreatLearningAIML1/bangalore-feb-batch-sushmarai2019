```python
#importing the required lib
import pandas as pd
import numpy as np
```

```python
#Reading from the csv file holding the new and old scheme data 
df = pd.read_csv("data.csv")
#renaming the 2nd and 3rd column
df=df.rename(columns = {'Old Scheme (in thousands)':'OldScheme'})
df=df.rename(columns = {'New Scheme (in thousands)':'NewScheme'})
print(df)
```

```python
# 1) Find the mean of old scheme and new scheme column
df.loc[:,["OldScheme","NewScheme"]].mean()
```

```python
# 2) Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs
# p_value < 0.05 => alternative hypothesis:
# null hypothesis: significance level of 0.05
#Checking if the mean of old and new are significantly different (5% level of significance)
from scipy.stats import ttest_1samp,ttest_ind,ttest_rel,t
t_statistic, p_value = ttest_ind(df.loc[:,"OldScheme"], df.loc[:,"NewScheme"])
print("p_value : ",p_value)
#print("we are accepting null hypothesis that the new scheme has significantly raised the output")
```

```python
# 2) Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs
#Assuming the expected value = 71.4 , which is equal to 5% of mean of old scheme
# p_value < 0.05 => alternative hypothesis:
# null hypothesis: significance level of 0.05
# one sample t-test and checking if the mean of new scheme is above 71.4
t_statistic, p_value = ttest_1samp(df.loc[:,"NewScheme"], 71.4)
print("p_value : ",p_value)
#print("we are accepting null hypothesis that the new scheme has significantly raised the output")
```
